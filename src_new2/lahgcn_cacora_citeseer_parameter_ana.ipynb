```python
import matplotlib.pyplot as plt

# 设置全局字体加粗和字号
plt.rcParams['font.family'] = 'Arial'  # 设置使用Arial字体
# plt.rcParams['font.weight'] = 'bold'   # 设置normal-不加粗;bold-加粗
plt.rcParams['font.size'] = 16

# 示例数据
x_values = [32, 64, 128, 256, 512]
y_values = [0.7695, 0.7848, 0.7735, 0.7744, 0.7759]

# # 创建折线图
# plt.plot(x_values, y_values, marker='o', linestyle='-')
# 创建折线图
plt.plot(range(len(x_values)), y_values, marker='o', linestyle='-')

# 添加标题和标签
plt.title('CA-Cora')
plt.xlabel('Different hidden size')
plt.ylabel('Accuracy')

# 设置横轴刻度为原始 x_values
plt.xticks(range(len(x_values)), x_values)


# 启用网格
plt.grid(True, linestyle='--')
# 显示图形
plt.show()
```
