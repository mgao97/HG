```python
import matplotlib.pyplot as plt

# 设置全局字体加粗和字号
plt.rcParams['font.family'] = 'Arial'  # 设置使用Arial字体
plt.rcParams['font.weight'] = 'bold'   # 设置normal-不加粗;bold-加粗
plt.rcParams['font.size'] = 40

# 示例数据
x_values = [32, 64, 128, 256, 512]
y_values = [0.4275, 0.4372, 0.4312, 0.4335, 0.4347]
plt.figure(figsize=(10, 8), dpi=300)
# # 创建折线图
# plt.plot(x_values, y_values, marker='o', linestyle='-')
# 创建折线图
plt.plot(range(len(x_values)), y_values, marker='o', linestyle='-')

# 添加标题和标签
plt.title('Citeseer',fontweight='bold',fontsize=40)
plt.xlabel('Different hidden size',fontweight='bold',fontsize=40)
plt.ylabel('Accuracy',fontweight='bold',fontsize=40)

# 设置横轴刻度为原始 x_values
plt.xticks(range(len(x_values)), x_values)


# 启用网格
plt.grid(True, linestyle='--')
# plt.savefig('citeseer_hidsize.pdf')

# 显示图形
plt.show()
```
